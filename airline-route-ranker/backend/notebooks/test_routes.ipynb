```python
import requests
import json
from pprint import pprint
```

```python
origin = "AMS"
destination = "LHE"

url = f"http://localhost:8000/api/rankings/{origin}/{destination}"

print(f"Fetching route rankings from {origin} to {destination}...")


response = requests.get(url, timeout=30)
response.raise_for_status()  # Raise exception for HTTP errors

data = response.json()
```

```python
print(json.dumps(data, indent=2))
```
